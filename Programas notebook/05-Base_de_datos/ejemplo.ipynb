## Matplotlib's two interfaces

### Imperative interface

```python
x = np.linspace(0,2,100)
plt.figure(figsize=(8,6))

plt.plot(x,x, label='Lineal', color='#c0000f', linewidth= 1.5, linestyle='-.')
plt.plot(x,np.cos(x),color='#c000ff',label='cosine', linewidth=2, linestyle=':')
plt.plot(x,np.sin(x),color='#c0d860',label='cosine', linewidth=2,)
plt.xlabel('X Label', size=14)
plt.ylabel('Y Label', size=14)
plt.title('Grafica de Linea', size=20)
plt.legend(loc='upper right')
```
